```python
#drop rows of null value
tmdb_movies.dropna(inplace=True)
tmdb_movies.shape
```

```python
#get the genre name in correct way
from ast import literal_eval
#if there is no values in genre then store[] in that place
tmdb_movies['genres']=tmdb_movies['genres'].fillna('[]')
#converting string representation of list to list of dictionaries
def safe_literal_eval(val):
    try:
        return literal_eval(val)
    except (ValueError, SyntaxError):
        return []

tmdb_movies['genres'] = tmdb_movies['genres'].apply(safe_literal_eval)

#extract value of name key in the genres dictionary
tmdb_movies['genres']=tmdb_movies['genres'].apply(lambda x: [i['name'] for i in x] if isinstance(x, list) else [])
tmdb_movies['genres']
```

```python
#simple recommendation system based on user rating
#1. Top movies based on rating
#V
vote_counts = tmdb_movies[tmdb_movies['vote_count'].notnull()]['vote_count'].astype('int')
```

```python
#r
vote_averages=tmdb_movies[tmdb_movies['vote_average'].notnull()]['vote_average'].astype('int')
```

```python
#C
C=vote_averages.mean()
C
```

```python
#minimum number of votes to be in our chart
m=vote_counts.quantile(0.95)
m
```

```python
#calculate weighted rating
def wr(x):
    v=x['vote_count']
    R=x['vote_average']
    return (v/(v+m)*R)+ (m/(m+v) * C)
```

```python
qualified=tmdb_movies[(tmdb_movies['vote_count']>=m) & (tmdb_movies['vote_count'].notnull()) & (tmdb_movies['vote_average'].notnull())][['title',
                                                                                                                                         'release_date',
                                                                                                                                         'genres','popularity',
                                                                                                                                       'runtime','vote_count',
                                                                                                                                        'vote_average']]
qualified['vote_average']=qualified['vote_average'].astype('int') #vote_average in in float in tmdb_movies dataset
qualified.shape
```

```python
qualified['wr']=qualified.apply(wr,axis=1)
qualified=qualified.sort_values(by='wr',ascending=False) #wr is more for for ratings so arranging in descending order
```

```python
#Top rated films based on specific genre
#make rating count cutoff 80%
#create list of specific type 
#converting the genre of each row to series
genre_list=tmdb_movies.apply(lambda x: pd.Series(x['genres']),axis=1)
genre_list
```

```python
#now stack is called to store all the genre in single column using multi index
genre_list=genre_list.stack()
genre_list
```

```python
genre_list=genre_list.reset_index(level=1,drop=True)
genre_list.name='genre' 
genre_list
```

```python
#creating new dataframe that drops previous genre coulumn and add new genre column which have one value in each row
genre_new=tmdb_movies.drop('genres',axis=1).join(genre_list)
genre_new.head().T
```

```python
#recommendation method for each type of genre
def genre_chart(genre,percentile=0.80):
    genre_df=genre_new[genre_new['genre']==genre]
    vote_counts=genre_df[genre_df['vote_count'].notnull()]['vote_count'].astype('int')
    vote_averages=genre_df[genre_df['vote_average'].notnull()]['vote_average'].astype('int')
    C=vote_averages.mean()
    m=vote_counts.quantile(percentile)
    genre_qualified=genre_df[(genre_df['vote_count']>=m) & (genre_df['vote_count'].notnull()) & (genre_df['vote_average'].notnull())][['title','genre','release_date','runtime','vote_count','vote_average','popularity']]
    genre_qualified['vote_average']=genre_qualified['vote_average'].astype('int')
    genre_qualified['wr']=genre_qualified.apply(lambda x:(x['vote_count']/(x['vote_count']+m) * x['vote_average']) + (m/(m+x['vote_count']) * C),
                                               axis=1)
    genre_qualified=genre_qualified.sort_values('wr',ascending=False).head(150)
    return genre_qualified
```

```python
#1. Based on description
tmdb_movies['overview']=tmdb_movies['overview'].fillna('')
tmdb_movies['description']=tmdb_movies['overview']+tmdb_movies['tagline']
tmdb_movies['description']=tmdb_movies['description'].fillna('')
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tf=TfidfVectorizer(analyzer='word',ngram_range=(1,2),min_df=0,stop_words='english')
tf_matrix=tf.fit_transform(tmdb_movies['description'])
```

```python
tf_matrix.shape
```

```python
from sklearn.metrics.pairwise import linear_kernel
cosine_sim=linear_kernel(tf_matrix,tf_matrix)
```

```python
cosine_sim[0]
```

```python
tmdb_movies=tmdb_movies.reset_index()
titles=tmdb_movies['title']
indices=pd.Series(tmdb_movies.index,index=tmdb_movies['title'])
indices.head()
```

```python
def description_recommendation(movie_name,n_rec=10):
    movieidx=indices[movie_name]
    rec_mov=list(enumerate(cosine_sim[movieidx]))
    rec_mov=sorted(rec_mov,key=lambda x: x[1],reverse=True)
    rec_mov=rec_mov[1:(n_rec+1)]
    rec_mov_indices=[i[0] for i in rec_mov]
    return titles.iloc[rec_mov_indices]
    
```

```python
cosine[0]
```

```python
cast_list=cast_list.reset_index()
titles=cast_list[['title','wr']]
indices=pd.Series(cast_list.index,index=cast_list['title'])
indices.head()
```

```python
def cast_recommendation_movie(movie_title,n_rec=10):
    movie_title = movie_title.lower()
    indices.index = indices.index.str.lower()
    input_movieid=indices[movie_title]
    rec_mov_cast=list(enumerate(cosine[input_movieid]))
    rec_mov_cast=sorted(rec_mov_cast,key=lambda x: x[1],reverse=True)
    rec_mov_cast=rec_mov_cast[1:(n_rec+1)]
    rec_mov_cast_indices=[i[0] for i in rec_mov_cast]
    return titles.iloc[rec_mov_cast_indices].sort_values(by='wr',ascending=False)
```

```python
#Content Based Recommendation more advanced based on cast,director,keywords,genres.
new_ds=tmdb_movies.iloc[:,0:18]
new_ds.rename(columns={'id':'movie_id'},inplace=True)
new_ds.head(2)
```

```python
content_rec_dataset=tmdb_credits.merge(new_ds,on='movie_id')
content_rec_dataset.head(2)
```

```python
content_rec_dataset['cast']=content_rec_dataset['cast'].apply(literal_eval)
content_rec_dataset['cast'].head()
```

```python
content_rec_dataset['crew']=content_rec_dataset['crew'].apply(literal_eval)
content_rec_dataset['crew'].head()
```

```python
content_rec_dataset['keywords']=content_rec_dataset['keywords'].apply(literal_eval)
content_rec_dataset['keywords'].head()
```

```python
content_rec_dataset['cast']=content_rec_dataset['cast'].apply(lambda x:[i['name'] for i in x] if isinstance(x,list) else '[]')
content_rec_dataset['cast']=content_rec_dataset['cast'].apply(lambda x : [str.lower(i.replace(' ','')) for i in x])
content_rec_dataset['cast']=content_rec_dataset['cast'].apply(lambda x : [i.replace('-','') for i in x])
content_rec_dataset['cast'].head()
```

```python
def director(x):
    for i in x:
        if i['job']=='Director':
         return i['name']
    return np.nan
content_rec_dataset['director']=content_rec_dataset['crew'].apply(director)
content_rec_dataset['director'] = content_rec_dataset['director'].astype('str').apply(lambda x: str.lower(x.replace(" ","")))
content_rec_dataset['director']=content_rec_dataset['director'].apply(lambda x: [x,x, x])
content_rec_dataset['director'].head(2)
```

```python
content_rec_dataset['keywords']=content_rec_dataset['keywords'].apply(lambda x: [i['name'] for i in x] if isinstance(x,list) else [])
content_rec_dataset['keywords'].head()
```

```python
keyword_stack = content_rec_dataset.apply(lambda x: pd.Series(x['keywords']), axis=1)
keyword_stack=keyword_stack.stack().reset_index(level=1,drop=True)
keyword_stack.name='keyword'
keyword_stack=keyword_stack.value_counts()
keyword_stack[:5]
```

```python
keyword_stack=keyword_stack[keyword_stack>1]
keyword_stack
```

```python
def keywords_sorted(x):
    words=[]
    for i in x:
        if i in keyword_stack:
            words.append(i)
    return words
```

```python
content_rec_dataset['keywords']=content_rec_dataset['keywords'].apply(keywords_sorted)
content_rec_dataset['keywords'].head()
```

```python
from nltk.stem.snowball import SnowballStemmer
stemmer=SnowballStemmer('english')
content_rec_dataset['keywords']=content_rec_dataset['keywords'].apply(lambda x:[stemmer.stem(i) for i in x])
content_rec_dataset['keywords']=content_rec_dataset['keywords'].apply(lambda x: [str.lower(i.replace(' ','')) for i in x])
```

```python
content_rec_dataset.head(2)
```

```python
#content recommendation based on cast,director,keywords,genres.

content_rec_dataset['cdkg']=content_rec_dataset['cast'] + content_rec_dataset['director'] + content_rec_dataset['keywords'] + content_rec_dataset['genres']
content_rec_dataset['cdkg'] = content_rec_dataset['cdkg'].apply(lambda x: ' '.join(x))
content_rec_dataset['cdkg'].head(2)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
count=CountVectorizer(analyzer='word',ngram_range=(1, 2),min_df=0, stop_words='english')
count_matrix=count.fit_transform(content_rec_dataset['cdkg'])
```

```python
from sklearn.metrics.pairwise import cosine_similarity
cosine_sim=cosine_similarity(count_matrix, count_matrix)
```

```python
content_rec_dataset=content_rec_dataset.reset_index()
```

```python
titles=content_rec_dataset['title']
indices = pd.Series(content_rec_dataset.index, index=content_rec_dataset['title'])
```

```python
def content_recommendation(movie_name,n_rec=10):
    movieidx=indices[movie_name]
    rec_mov=list(enumerate(cosine_sim[movieidx]))
    rec_mov=sorted(rec_mov,key=lambda x: x[1],reverse=True)
    rec_mov=rec_mov[1:(n_rec+1)]
    rec_mov_indices=[i[0] for i in rec_mov]
    return titles.iloc[rec_mov_indices]
```
